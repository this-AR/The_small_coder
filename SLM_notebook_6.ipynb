```python
import requests, json, time, os, re
from tqdm.auto import tqdm

# --- 1. SEGMENT CONFIGURATION ---
MEMBER_ID = 0  # 👈 CHANGE THIS (0 to 9) for each group member
API_KEY = "AIzaSyAcQY3tmM9pB7LvzfBtDy1qb83TOTKrBvw"
INPUT_FILE = "/content/drive/MyDrive/SLM_DATA/stage2_train.jsonl"
OUTPUT_FILE = f"gemma_distilled_part_{MEMBER_ID + 1}.jsonl"

# --- 2. EXPERT COACH PROMPT (Strict Syntax) ---
BRIDGE_PROMPT = r"""
You are an expert Competitive Programming Coach. Transform this reasoning trace into a High-Density Logical Blueprint.

### INSTRUCTIONS:
1. [DIFFICULTY]: EASY, MEDIUM, or HARD.
2. [TOPICS]: 3-4 precise algorithmic tags.
3. [GOAL]: Define the Win Condition in one crisp sentence.
4. [LOGIC]: Extract core mathematical 'pivot' or logical point that explain the core logic of the problem.
5. [PSEUDOCODE]: Write syntactically valid Python logic. Avoid fluff.
6. [CORNER CASES]: List 1-2 critical edge cases.

### FORMATTING RULE:
- Tone: Technical and objective.
- No bolding (**) inside the values.
- Input Reasoning: {reasoning}

### YOUR DENSE OUTPUT:
"""

def validate_response(text):
    required = ["[DIFFICULTY]", "[TOPICS]", "[GOAL]", "[LOGIC]", "[PSEUDOCODE]", "[CORNER CASES]"]
    missing = [tag for tag in required if tag not in text]
    return (True, []) if not missing else (False, missing)

def call_gemma_with_reporting(prompt):
    url = f"https://generativelanguage.googleapis.com/v1beta/models/gemma-3-27b-it:generateContent?key={API_KEY}"
    try:
        response = requests.post(url, json={"contents": [{"parts": [{"text": prompt}]}],
                                     "generationConfig": {"temperature": 0.1, "maxOutputTokens": 4096}}, timeout=120)
        if response.status_code == 200:
            return response.json()['candidates'][0]['content']['parts'][0]['text']
        elif response.status_code == 429:
            print("\n⏳ Quota Hit. Waiting 60s...")
            time.sleep(60)
            return call_gemma_with_reporting(prompt)
        return None
    except Exception as e:
        print(f"\n❌ Connection Error: {str(e)}")
        return None

def run_distillation():
    if not os.path.exists(INPUT_FILE):
        print(f"❌ Error: {INPUT_FILE} not found. Mount Drive first.")
        return

    with open(INPUT_FILE, 'r') as f:
        all_lines = f.readlines()

    # Divide 2000 questions into 10 segments of ~200
    chunk_size = len(all_lines) // 10
    start = MEMBER_ID * chunk_size
    end = start + chunk_size if MEMBER_ID < 9 else len(all_lines)
    my_chunk = all_lines[start:end]

    print(f"📡 Member {MEMBER_ID} | Range: {start}-{end} | Output: {OUTPUT_FILE}")

    with open(OUTPUT_FILE, 'a', buffering=1) as out:
        for i, line in enumerate(tqdm(my_chunk, desc=f"Member {MEMBER_ID}")):
            try:
                data = json.loads(line)
                raw_out = data.get('output', '')

                # API Call
                res_text = call_gemma_with_reporting(BRIDGE_PROMPT.format(reasoning=raw_out[:1800]))

                if res_text:
                    # 1. Validation
                    is_valid, missing_tags = validate_response(res_text)
                    if not is_valid:
                        print(f"\n⚠️ Skip Row {start+i}: Missing {missing_tags}")
                        continue

                    # 2. Extract Clean Code (Format Fixer)
                    code_match = re.search(r'<code>(.*?)</code>', raw_out, re.DOTALL)
                    clean_code = code_match.group(1).strip() if code_match else ""
                    if not clean_code: continue

                    # 3. Parse Metadata
                    diff = re.search(r"\[DIFFICULTY\]:?\s*(.*)", res_text, re.I)
                    topi = re.search(r"\[TOPICS\]:?\s*(.*)", res_text, re.I)
                    d_val = diff.group(1).split('\n')[0].strip() if diff else "MEDIUM"
                    t_val = topi.group(1).split('\n')[0].strip() if topi else "ALGORITHM"

                    # 4. Save Final Clean Architecture
                    final_data = {
                        "instruction": data.get('instruction', ''),
                        "output": f"<tags> {d_val}, {t_val} </tags>\n<think>\n{res_text}\n</think>\n<code>\n{clean_code}\n</code>"
                    }
                    out.write(json.dumps(final_data) + '\n')
                    os.fsync(out.fileno())

                # 🔥 30s Delay for 2 QPM (Safe for 15k TPM)
                time.sleep(30)
            except: continue

run_distillation()
```
